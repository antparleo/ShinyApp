# Phylotypes

```python
# Libraries
import pandas as pd
import numpy as np
```

## Training

```python
phylo_train = pd.read_csv('../Training_data/phylotypes/phylotype_relabd.5e_1.csv')
phylo_train
```

## Test (PTB and early PTB)

```python
phylo_test_preterm = pd.read_csv('../Validation_data/PTB/phylotypes/phylotype_relabd.5e_1.csv')
phylo_test_preterm
```

```python
phylo_test_early = pd.read_csv('../Validation_data/early_PTB/phylotypes/phylotype_relabd.5e_1.csv')
phylo_test_early
```

## Merge all of them

```python
phylo_all = pd.concat([phylo_train, phylo_test_preterm, phylo_test_early]).drop_duplicates()
phylo_all.specimen = phylo_all.specimen.replace({'K': 'W', 'V': 'S'}, regex = True)
phylo_all
```

## Save data

```python
phylo_all.to_csv('../Results/phylotypes_all.csv', index=False)
```
