# Create suitable file for CST alluvial plot

```python
import pandas as pd
```

## Load data

### Training data

```python
cst_train = pd.read_csv('../Training_data/community_state_types/cst_valencia.csv')
cst_train
```

### Validation data

```python
cst_validation_preterm = pd.read_csv('../Validation_data/PTB/community_state_types/cst_valencia.csv')
cst_validation_early = pd.read_csv('../Validation_data/early_PTB//community_state_types/cst_valencia.csv')
```

```python
cst_validation = pd.concat([cst_validation_preterm,cst_validation_early], axis=0).drop_duplicates()
cst_validation.shape
```

## Create table with all information

```python
cst_all = pd.concat([cst_train, cst_validation], axis = 0)
cst_all.specimen = cst_all.specimen.replace({'K': 'W', 'V': 'S'}, regex = True)
cst_all
```

## Save date

```python
cst_all.to_csv('../Results/cst_all.csv', index= False)
```

## Suitable format for alluvial plot

```python
metadata_all = pd.read_csv('../Results/metadata_all.csv')
specimenCST = cst_all.specimen
metadata = metadata_all.query('specimen in @specimenCST')
metadata
```

```python
CST_metadata = pd.merge(metadata, cst_all, on = 'specimen')
CST_metadata
```

```python
# Participant with at least one samples in different trimesters.

dup = {}
to_keep = []
for p in pd.unique(CST_metadata.participant_id):
    dup[p] = pd.unique(CST_metadata.loc[CST_metadata.participant_id == p,"Trimester"])
    if len(dup[p]) > 1:
        to_keep.append(p)

print(len(to_keep))
cst_2_trim = CST_metadata.query('participant_id in @to_keep')
```

```python
my_list = []
for t in range(1,4):
    cst_trimester = cst_2_trim.query('Trimester == @t')

    dup = {}
    
    for index,row in cst_trimester.iterrows():
        if row.participant_id not in dup.keys():
            dup[row.participant_id] = (row.collect_wk,row["specimen"])
        else:
            if row.collect_wk > dup[row.participant_id][0]:
                dup[row.participant_id] = (row.collect_wk, row["specimen"])

    my_index = [i[1] for i in dup.values()] 
    cst_trimester_final = cst_trimester.query('specimen in @my_index')
    my_list.append(cst_trimester_final)

cst_alluvia = pd.concat(my_list,axis=0).sort_index()
cst_alluvia
```
