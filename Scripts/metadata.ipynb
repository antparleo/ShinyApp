# Adjust Metdata for Shiny

```python
# Libraries 

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Metadata training

```python
metadata_train = pd.read_csv('../Training_data/metadata/metadata.csv')
metadata_train.drop('was_preterm',axis = 1, inplace = True)
```

```python
metadata_train[['participant_id','was_term','was_early_preterm','project']].drop_duplicates()
to_check = ['D00061','D00028','D00029','D00059']
metadata_train.query('participant_id in @to_check')
```

```python
metadata_test = pd.read_csv('../Validation_data/key/metadata_full.csv')
metadata_test.project = metadata_test.project.replace({'K': 'W', 'V': 'S'})
metadata_test.specimen = metadata_test.specimen.replace({'K': 'W', 'V': 'S'}, regex = True)
metadata_test.participant_id = metadata_test.participant_id.replace({'K': 'W', 'V': 'S'}, regex = True)
metadata_test
```

```python
early_preterm = pd.read_csv('../Validation_data/key/key_early_preterm.csv')
early_preterm.was_early_preterm = early_preterm.was_early_preterm.replace({1: True, 0: False})
metadata_test['was_early_preterm'] = metadata_test.participant_id.map(dict(zip(early_preterm.participant_id,early_preterm.was_early_preterm)))
```

```python
metadata_test[['participant_id','was_term','was_early_preterm','project']].drop_duplicates().project.value_counts()
```

```python
metadata = pd.concat([metadata_train, metadata_test], axis=0)

metadata = metadata.replace({'Above_38' : 'Above 29', 'Below_18' : 'Below 28',
                 '18_to_28' : 'Below 28', '29-38':'Above 29'})

metadata = metadata.replace({'American Indian or Alaska Native' : 'Other', 'Native Hawaiian or Other Pacific Islander' : 'Other', 'Asian':'Other'})

display(metadata)
metadata.age.value_counts().sort_values(ascending = True)
```

### Create a new colunm for age

```python
age_ranges = ['Unknown','Above 29', 'Below 28']
range_cat = metadata.query('age in @age_ranges')
range_cat['Range'] = range_cat.age
range_cat.Range.value_counts()
```

```python
range_num = metadata.query('age not in @age_ranges')
range_num.loc[:,'age'] = range_num['age'].astype(float)
range_num['Range'] = pd.cut(range_num.age, bins = [0,28,200], labels= ['Below 28','Above 29'])
range_num.Range.value_counts()
```

```python
metadata_all = pd.concat([range_cat,range_num], axis=0)
display(metadata_all)
metadata_all.Range.value_counts()
```

It is considered as early preterm birth before 32 weeks (not included) and as preterm birth before 37 weeks (not included).

```python
metadata_all['Type'] = pd.cut(metadata_all['delivery_wk'],bins=[0,31,36,200], labels= ['early', 'preterm','term'])
metadata_all['Trimester'] = pd.cut(metadata_all.collect_wk,[0,13,26,42],labels=[1,2,3])
metadata_all = metadata_all.rename(columns={'NIH Racial Category':'Race', 'Range' : 'Age'})
metadata_all
```
