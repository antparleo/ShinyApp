```python
ptw_1e_1_validation = pd.concat([ptw_1e_1_validation_early, ptw_1e_1_validation_preterm]).drop_duplicates()
```

```python
ptw_1e_1 = pd.concat([ptw_1e_1_training, ptw_1e_1_validation])
print(ptw_1e_1.shape)
ptw_1e_1.specimen = ptw_1e_1.specimen.replace({'K': 'W', 'V': 'S'}, regex = True)
ptw_1e_1.columns = ptw_1e_1.columns.map(lambda pt: pt.replace('pt__', 'pt__vaginal.202208__0.1____'))
ptw_1e_1.to_csv('../Results/phylotypes_1e1_heatmap.csv', index=False)
```

```python
ptw_1e_1.columns.duplicated().sum()
```

```python
sv_tax = pd.read_csv(
    "../Training_data/taxonomy/sv_taxonomy.csv"
)
sv_pt1e_1 = {
    r.sv: r.phylotype.replace('pt__', 'pt__vaginal.202208__0.1____')
    for (i, r) in
    pd.read_csv(
        "../Training_data/phylotypes/pt.1e-1.csv"
    ).iterrows()
}
sv_pt1e_1
```

```python
sv_tax['pt1e_1'] = sv_tax.sv.apply(sv_pt1e_1.get)
sv_tax
```

```python
pt_sp = {
    pt: ptt.tax_name
    for pt, ptt in
    sv_tax[
        sv_tax['rank'] == 'species'
    ].groupby([
        'pt1e_1', 'tax_name'
    ]).count().sort_values(
        'sv',
        ascending=False
    ).reset_index().groupby(
        'pt1e_1'
    ).first().iterrows()
}
pt_sp
```

```python
ptnosp = set(ptw_1e_1.columns) - set(pt_sp.keys())

pt_sp.update({
    pt: ptt.tax_name+" spp."
    for pt, ptt in
    sv_tax[
        (sv_tax['want_rank'] == 'species') &
        (sv_tax.pt1e_1.apply(lambda pt: pt in ptnosp))
    ].groupby([
        'pt1e_1', 'tax_name'
    ]).count().sort_values(
        'sv',
        ascending=False
    ).reset_index().groupby(
        'pt1e_1'
    ).first().iterrows()
})
pt_sp
```

```python
phylo_specie = pd.DataFrame.from_dict(pt_sp, orient = 'index').reset_index().rename({'index':'phylotypes', 0:'Specie'}, axis = 1)
phylo_specie.to_csv('../Results/phylo_specie_1e1.csv', index = False)
```
