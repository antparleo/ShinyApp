# Diversity

```python
import pandas as pd
```

## Training

```python
diver_training = pd.read_csv('../Training_data/alpha_diversity/alpha_diversity.csv')
diver_training
```

```python
diver_validation_preterm = pd.read_csv('../Validation_data/PTB/alpha_diversity/alpha_diversity.csv')
diver_validation_preterm.query('specimen == "K00001-03"')
```

```python
diver_validation_early = pd.read_csv('../Validation_data/early_PTB/alpha_diversity/alpha_diversity.csv')
diver_validation_early.query('specimen == "K00001-03"') # There are duplicated
```

```python
diver_validation = pd.concat([diver_validation_preterm,diver_validation_early], axis=0).drop_duplicates()
len(diver_validation.specimen.apply(lambda x: x[0:-3]).drop_duplicates())
```

## Create diver all table

```python
diver_all = pd.concat([diver_training, diver_validation], axis = 0)
diver_all.specimen = diver_all.specimen.replace({'K': 'W', 'V': 'S'}, regex = True)
diver_all
```

We add Participant column

```python
diver_all['participant_id'] = diver_all.specimen.apply(lambda x: x[0:-3])
diver_all
```

## Save data

```python
diver_all.to_csv('../Results/diversity_all.csv', index= False)
```
